```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# MNIST Dataset

```python
data = pd.read_csv("/Users/layla/Documents/mnist.csv")
data
```

# Data Preprocessing

```python
#We want to predict the label
data_copy = data.copy()
class_names = data_copy['label'].unique()
class_names
```

```python
y = data_copy['label']
y
```

```python
X = data_copy.drop('label', axis = 1)
X
```

# Train, Test and Validation

```python
pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#ID = 900202391
k = np.sqrt(2391)
k = int(k)
k
```

# KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = k)
```

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

```python
pred = knn.predict(X_test)
pred
```

```python
from sklearn.model_selection import cross_validate
```

```python
metrics = ['accuracy', 'precision_weighted', 'recall_weighted', 'f1_weighted', 'roc_auc_ovr']
cross_val = cross_validate(knn, X, y, cv = 10, scoring = metrics)
```

```python
print("\naccuracy:\n")
print(cross_val['test_accuracy'])
print("_______________________________________________________________\n")
print("precision:\n")
print(cross_val['test_precision_weighted'])
print("_______________________________________________________________\n")
print("recall:\n")
print(cross_val['test_recall_weighted'])
print("_______________________________________________________________\n")
print("f1:\n")
print(cross_val['test_f1_weighted'])
print("_______________________________________________________________\n")
print("AUC:\n")
print(cross_val['test_roc_auc_ovr'])
```

```python
k_range = list(range(1,10))
param_grid = {'n_neighbors': k_range} 
knn2 =  KNeighborsClassifier()
```

```python
find_k = GridSearchCV(knn2, param_grid, cv = 10, scoring = 'accuracy')
find_k.fit(X_train, y_train)
```
